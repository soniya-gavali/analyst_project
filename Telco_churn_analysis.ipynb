```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=pd.read_csv('Telco-Customer-Churn.csv')
df.head()
```

```python
df.info()
```

#replacing blanks with 0 as tenure is 0 and no total charges are recorded

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ","0")
df["TotalCharges"]=df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df["customerID"].duplicated().sum()
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

#converted 0 and 1 values of senior Citizen to yes/no to make it easier to understand ->this is of 15th cell.

```python
df.head()
```

```python
ax=sns.countplot(x = 'Churn',data = df)
ax.bar_label(ax.containers[0])
```

```python
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'],labels = gb.index,autopct = "%1.2f%%")
```

#from the given pie chart we can conclude that 26.54% of our customers have churned out. #not let's explore the reason behind it

```python
plt.figure(figsize = (4,4))
sns.countplot(x = "gender",data = df, hue = "Churn")
plt.title("Churn by Gender")
```

```python
plt.figure(figsize = (4,5))
ax = sns.countplot(x = "SeniorCitizen",data = df)
ax.bar_label(ax.containers[0])
plt.title("count of customers by siniorcitizen")
```

#comparative a greater percentage of people in senior citizen catagory have churned

```python
plt.figure(figsize = (4,4))
sns.countplot(x = "SeniorCitizen",data = df, hue = "Churn")
plt.title("Churn by SeniorCitizen")
```

```python
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

fig, ax = plt.subplots(figsize=(9,6))

total_counts.plot(kind='bar',stacked=True, ax=ax,color=['#1f77b4', '#ff7f0e'])

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x,y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center')

plt.title('Churn by senior citizen (Stacked Bar Chart)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Churn',loc='upper right')

plt.show()
```

```python
plt.figure(figsize = (9,4))
sns.histplot(x = "tenure", data = df, bins=50, hue = "Churn")
plt.show()
```

#people who hava used our services for a long time have stayed and people who have used our services #1 or 2 months have churned

```python
plt.figure(figsize = (4,5))
ax = sns.countplot(x = "Contract", data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of customers by Contract")
```

#people who have month to month contract are likely to churn then from from those who have 1 or 2 years or contract

```python
df.columns.values
```

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15,n_rows * 4))

axes = axes.flatten()

for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], hue = df["Churn"])
    axes[i].set_title(f'Count plot of (col)')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
    
```

```python
plt.figure(figsize = (10,4))
ax = sns.countplot(x = "PaymentMethod",data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("churned customers by payment method")
```
